```python
import cv2
```

```python
img = cv2.imread("data/3.jpg")
print(img.shape)
print(img)
img = cv2.resize(img,(500, 500))
cv2.imshow("Resized Image", img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
